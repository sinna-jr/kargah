```python
received_msg = "recived_msg.txt"
sent_msg = "sent_msg.txt"
```

```python
with open(received_msg, "r", encoding = "utf-8") as file:
    file_contents = file.read()
    content = file_contents.split("\n")
    name = ""
    for i in range(len(content)):
        if "First Name" in content[i]:
            name = content[i+1]

print(file_contents)
```

```python
with open(sent_msg, "r", encoding = "utf-8") as file:
    file_contents = file.read()
    file_contents = file_contents.replace("NAME" , name)
print(file_contents)
```
