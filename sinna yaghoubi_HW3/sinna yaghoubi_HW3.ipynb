```python
import numpy as np
import matplotlib.pyplot as plt

x = np.linspace(-1, 1, 100)
y = np.sin(x)

plt.plot(x , y , 'b' , linewidth=2)

plt.title("sin(x)")
plt.xlabel("x")
plt.ylabel("y")
plt.legend(["sin(x)"])
plt.grid()

plt.savefig("img1_1.png" , dpi=100)
plt.savefig("img1_2.jpg" , dpi=300)

plt.show()
```

```python
import numpy as np
import matplotlib.pyplot as plt

x = np.linspace(-1, 1, 100)
y = 2*np.sin(x/2) * (np.cos(x/4)**2) - 2*np.sin(x/2) * (np.sin(x/4)**2)

plt.plot(x , y , 'r', linewidth=2)

plt.title("function 2")
plt.xlabel("x")
plt.ylabel("y")
plt.legend(["f(x)"])
plt.grid()

plt.savefig("img2_1.png" , dpi=100)
plt.savefig("img2_2.jpg" , dpi=300)

plt.show()
```

```python
import numpy as np
import matplotlib.pyplot as plt

x = np.linspace(-1 , 1 , 100)

y1 = np.sin(x)
y2 = 2*np.sin(x/2) * (np.cos(x/4)**2) - 2*np.sin(x/2) * (np.sin(x/4)**2)

plt.plot(x , y1 , 'b', linewidth=2)
plt.plot(x , y2 , 'r', linewidth=2)

plt.title("combine")
plt.xlabel("x")
plt.ylabel("y")

plt.legend(["sin(x)" , "f(x)"])
plt.grid()

plt.savefig("img3.png" , dpi=300)

plt.show()
```
