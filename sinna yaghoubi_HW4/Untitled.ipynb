```python
import cv2
import matplotlib.pyplot as plt

img = cv2.imread("img.png")
gray = cv2.cvtColor(img, cv2.COLOR_BGR2GRAY)

hist = cv2.calcHist([gray],[0],None,[256],[0,256])

img2 = gray.copy()
rows , cols = gray.shape

for i in range(rows):
    for j in range(cols):
        if gray[i,j] == 255:
            img2[i,j] = 0
        elif gray[i,j] == 0:
            img2[i,j] = 255

hist2 = cv2.calcHist([img2],[0],None,[256],[0,256])

img3 = 255 - gray
hist3 = hist[::-1]

plt.subplot(3,2,1)
plt.imshow(gray, cmap='gray')

plt.subplot(3,2,2)
plt.plot(hist)

plt.subplot(3,2,3)
plt.imshow(img2, cmap='gray')

plt.subplot(3,2,4)
plt.plot(hist2)

plt.subplot(3,2,5)
plt.imshow(img3, cmap='gray')

plt.subplot(3,2,6)
plt.plot(hist3)


plt.show()
```
